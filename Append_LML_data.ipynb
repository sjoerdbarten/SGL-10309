```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import isfile, join

pd.set_option('display.max_rows', 50)
```

```python
#DATA FROM https://data.rivm.nl/data/luchtmeetnet/

datapath_NO = '2013-2023/LML/NO/'
datapath_NO2 = '2013-2023/LML/NO2/'
datapath_NH3 = '2013-2023/LML/NH3/'

#NL10131 = Vredepeel
#NL10633 = Zegveld
#NL10738 = Wekerom

df = pd.DataFrame()
df_NO = pd.concat([pd.read_csv(datapath_NO+f, skiprows=9, header=0, sep=';', encoding= 'unicode_escape') 
                   for f in listdir(datapath_NO) if isfile(join(datapath_NO, f))], axis=0)
df_NO2 = pd.concat([pd.read_csv(datapath_NO2+f, skiprows=9, header=0, sep=';', encoding= 'unicode_escape') 
                    for f in listdir(datapath_NO2) if isfile(join(datapath_NO2, f))], axis=0)
#df_NH3 = pd.concat([pd.read_csv(datapath_NH3+f, skiprows=15, header=0, sep=';', encoding= 'unicode_escape') 
#                    for f in listdir(datapath_NH3) if isfile(join(datapath_NH3, f))], axis=0)
df.index = pd.to_datetime(df_NO[' Begindatumtijd'],format="%Y%m%d %H:%M")
df['date-time'] = df.index.strftime('%Y-%m-%d %#H:%M:%S')
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['hour'] = df.index.hour

#df_NH3["NH3_Vredepeel"] = df_NH3[["NL10131"]].sum(axis=1, min_count=1)
df_NO["NO_Vredepeel"] = df_NO[["NL10131"]].sum(axis=1, min_count=1)
df_NO2["NO2_Vredepeel"] = df_NO2[["NL10131"]].sum(axis=1, min_count=1)
#df_NH3["NH3_Zegveld"] = df_NH3[["NL10633"]].sum(axis=1, min_count=1)
df_NO['NO_Zegveld'] = df_NO[["NL10633"]].sum(axis=1, min_count=1)
df_NO2['NO2_Zegveld'] = df_NO2[["NL10633"]].sum(axis=1, min_count=1)
#df_NH3["NH3_Wekerom"] = df_NH3[["NL10738"]].sum(axis=1, min_count=1)
df_NO['NO_Wekerom'] = df_NO[["NL10738"]].sum(axis=1, min_count=1)
df_NO2['NO2_Wekerom'] = df_NO2[["NL10738"]].sum(axis=1, min_count=1)

#df["NH3_Vredepeel"] = df_NH3["NH3_Vredepeel"].values
df["NO_Vredepeel"] = df_NO["NO_Vredepeel"].values
df["NO2_Vredepeel"] = df_NO2["NO2_Vredepeel"].values
#df["NH3_Zegveld"] = df_NH3["NH3_Zegveld"].values
df['NO_Zegveld'] = df_NO["NO_Zegveld"].values
df['NO2_Zegveld'] = df_NO2["NO2_Zegveld"].values
#df["NH3_Wekerom"] = df_NH3["NH3_Wekerom"].values
df['NO_Wekerom'] = df_NO["NO_Wekerom"].values
df['NO2_Wekerom'] = df_NO2["NO2_Wekerom"].values

df = df.interpolate(limit_direction='both').reset_index().drop([' Begindatumtijd'],axis=1)

df.to_csv('2013-2023/Data_AQ_2023_python.csv', float_format='%.2f')

# Combining dataframes here

df_vredepeel_tofill = pd.read_csv('2013-2022/Data_AQ_Vredepeel_2013-2022.csv',sep=';',header=0)
df_zegveld_tofill = pd.read_csv('2013-2022/Data_AQ_Zegveld_2013-2022.csv',sep=';',header=0)
df_wekerom_tofill = pd.read_csv('2013-2022/Data_AQ_Wekerom_2013-2022.csv',sep=';',header=0)

df_new = df.copy()
#df_new["NH3"] = df_new["NH3_Vredepeel"]
df_new['NO'] = df_new['NO_Vredepeel']
df_new['NO2'] = df_new['NO2_Vredepeel']
df_new = df_new.drop(['NO_Vredepeel','NO2_Vredepeel','NO_Zegveld','NO2_Zegveld','NO_Wekerom','NO2_Wekerom'],axis=1)
#df_new = df_new.drop(['NH3_Vredepeel','NO_Vredepeel','NO2_Vredepeel',
#                      'NH3_Zegveld','NO_Zegveld','NO2_Zegveld',
#                      'NH3_Wekerom','NO_Wekerom','NO2_Wekerom'],axis=1)
df_vredepeel = pd.concat([df_vredepeel_tofill,df_new])
df_vredepeel.to_csv('2013-2023/Data_AQ_Vredepeel_2013-2023_noNH3.csv',index=False,sep=';', float_format='%.2f')

df_new = df.copy()
#df_new["NH3"] = df_new["NH3_Zegveld"]
df_new['NO'] = df_new['NO_Zegveld']
df_new['NO2'] = df_new['NO2_Zegveld']
df_new = df_new.drop(['NO_Vredepeel','NO2_Vredepeel','NO_Zegveld','NO2_Zegveld','NO_Wekerom','NO2_Wekerom'],axis=1)
#df_new = df_new.drop(['NH3_Vredepeel','NO_Vredepeel','NO2_Vredepeel',
#                      'NH3_Zegveld','NO_Zegveld','NO2_Zegveld',
#                      'NH3_Wekerom','NO_Wekerom','NO2_Wekerom'],axis=1)
df_zegveld = pd.concat([df_zegveld_tofill,df_new])
df_zegveld.to_csv('2013-2023/Data_AQ_Zegveld_2013-2023_noNH3.csv',index=False,sep=';', float_format='%.2f')

df_new = df.copy()
#df_new["NH3"] = df_new["NH3_Wekerom"]
df_new['NO'] = df_new['NO_Wekerom']
df_new['NO2'] = df_new['NO2_Wekerom']
df_new = df_new.drop(['NO_Vredepeel','NO2_Vredepeel','NO_Zegveld','NO2_Zegveld','NO_Wekerom','NO2_Wekerom'],axis=1)
#df_new = df_new.drop(['NH3_Vredepeel','NO_Vredepeel','NO2_Vredepeel',
#                      'NH3_Zegveld','NO_Zegveld','NO2_Zegveld',
#                      'NH3_Wekerom','NO_Wekerom','NO2_Wekerom'],axis=1)
df_wekerom = pd.concat([df_wekerom_tofill,df_new])
df_wekerom.to_csv('2013-2023/Data_AQ_Wekerom_2013-2023_noNH3.csv',index=False,sep=';', float_format='%.2f')
```
